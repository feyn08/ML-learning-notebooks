```python
print(f"The size of the dataset: {df.shape}")
```

```python
df.drop(['body'], axis = 1, inplace = True) #deleting a feature that has the highest number of missing values
```

```python
print(f"The size of the dataset after dropping a feature: {df.shape}")
```

```python
from sklearn.impute import SimpleImputer #2. Value Imputation: not deleting the feature but replacing the missing values in a column with a value using a statistically driven strategy(mean, median, mode)
```

```python
print(f"The no. of missing values in  age column before imputation: {df.age.isnull().sum()}")
```

```python
imp = SimpleImputer(strategy = 'mean')
df['age'] = imp.fit_transform(df[['age']])
```

```python
print(f"The no. of missing values in  age column after imputation: {df.age.isnull().sum()}")
```

```python
#3. if different columns have different datatypes: 
def get_parameters(df):
    parameters = {}
    for col in df.columns[df.isnull().any()]:
        if(df[col].dtype == 'float64' or df[col].dtype == 'int64' or df[col].dtype == 'int32'):
            strategy = 'mean'
        else:
            strategy = 'most_frequent'
        missing_values = df[col][df[col].isnull()].values[0]
        parameters[col] = {'missing_values': missing_values, 'strategy': strategy}
    return parameters
```

```python
parameters = get_parameters(df)
```

```python
for col, param in parameters.items():
    missing_values = param['missing_values']
    strategy = param['strategy']
    imp = SimpleImputer(missing_values = missing_values, strategy = strategy)
    df[col] = imp.fit_transform(df[[col]]).ravel()
```
